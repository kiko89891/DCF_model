```python
import numpy as np
import pandas as pd
from yahoofinancials import YahooFinancials
from functools import reduce
```

```python

# Define the stock symbol and the period for the financial data
stock_symbol = 'PII'  # Replace with the desired stock symbol'
length_period=4 # year behind to extract
# Create a YahooFinancials object
yahoo_financials = YahooFinancials(stock_symbol)
length_period_income=5 # the website has 5 years instead of 4 on data idk why
```

```python
# Initialize data
balance_sheet = yahoo_financials.get_financial_stmts('annual', 'balance')
balance_sheet_data = balance_sheet['balanceSheetHistory'][stock_symbol]
```

```python
cash_flow = yahoo_financials.get_financial_stmts('annual', 'cash')
cash_flow_data = cash_flow['cashflowStatementHistory'][stock_symbol]
```

```python
income_statement = yahoo_financials.get_financial_stmts('annual', 'income')
income_statement_data = income_statement['incomeStatementHistory'][stock_symbol]
```

```python
########## Function for extracation of data
```

```python
def extract(data,stock_symbol,period):
    frame=[]
    for i in range(0,period):
        x=data[i]      # take the n-th index. Convert it to dataframe and save it to frame. Then join(concat) the data
        x=pd.DataFrame(x)
        frame.append(x)
    sheet = pd.concat(frame, join='outer', axis=1)
    return sheet 
    
```
